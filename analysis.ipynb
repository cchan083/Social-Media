## Feature data investigation

What is the mean time spent on each video category?

```python
mean_time = df.groupby('Video Category')['Total Time Spent'].mean().reset_index()

plt.figure(figsize=(9,5))
plt.bar(mean_time['Video Category'], mean_time['Total Time Spent'], color='red')
plt.xlabel('Video Category')
plt.ylabel('Mean time spent')
plt.title('Mean time spent for each video category')
plt.xticks(fontsize=8)
plt.show()
```

```python
df['Video Category'].value_counts()
```

What Video Category is engaged with the most?

```python
df['Engagement'].describe()
```

Mean engagement for each Video category

```python
engagement_mean = df.groupby('Video Category')['Engagement'].mean().reset_index()

engagement_mean.plot(
    kind='barh',
    x='Video Category',
    y='Engagement',
    title='Mean Engagement scored for each video category'
)
```

Trends looks like it is the most engaged with, while ASMR and Gaming are the least engaged with.

But why is this?

We can filter out only the entries with ASMR and Gaming, and see their 'Reason'

```python
asmr_gaming = df.loc[
    (df['Video Category'] == 'ASMR')
    |
    (df['Video Category'] == 'Gaming')
]
reasons_counted = asmr_gaming['Watch Reason'].value_counts()
reasons_counted.plot(
    kind='pie',
    autopct='%1.2f%%',
    title='What is the reason for watching asmr or gaming?',
    ylabel=''
)
```

We can see clearly that most watchers of ASMR and Gaming watch it out of Habit, and boredom, with less than a third watching it for entertainment or procrastination. This could explain the reason why the engagement in the content is low compared to other types of content, and users watch mainly watch content to 'pass the time'

Watching Jokes/Memes, and pranks leads to the most productivity lost by far, but how does this affect their income?

```python
income_by_videos = df.groupby('Video Category')['Income'].mean().reset_index()
income_by_videos = income_by_videos.sort_values(by='Income')
plt.figure(figsize=(9,5))
plt.bar(income_by_videos['Video Category'], income_by_videos['Income'], color='green')
plt.xlabel('Video Category')
plt.ylabel('Income')
plt.title('Mean Income for each video category')
plt.xticks(fontsize=8)
plt.show()
```

Seems to not affect their income a lot as they are within the middle of all the Video Categories

### Is it true that younger people have less self control than older people?

```python
df['Age'].describe()
```

We can define young as being ages 18 to 29 as it is the Lower Quartile, and old as 52-64 as it is the upper quartile

Old people seem to have less self control towards social media than young people, this could be because they seem more things online about cybersickness

What is the distribution of Self Control towards ages?

```python
average_age = df.groupby('Self Control')['Age'].mean().reset_index().sort_values(by='Self Control')



plt.figure(figsize=(10, 6))
plt.bar(average_age['Self Control'], average_age['Age'], color='#3357FF')
plt.title('Average Age for each level of self control')
plt.xlabel('Self control level')
plt.ylabel('Average age')
plt.show()
```

It looks like that the lowest self control level has a mean age of around 45-47, but this has very little significance as it looks like most entries of data are aged around 40-45, and there is a even amount of entries at the Old and young ages, making the mean of each self control level around 40-45

```python
df['Age'].plot(
    kind='hist',
    x='Age',
    bins=30
)
```
